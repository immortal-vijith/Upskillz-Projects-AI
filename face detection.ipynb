```python
img=cv2.imread("myimage.jpeg")
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
gr_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gr_img)
```

```python
faces = face_cascade.detectMultiScale(gr_img, 1.3, 5,minSize=(30,30))
print(faces)
```

```python
image=cv2.imread("myimage.jpeg")
for(x,y,w,h)in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),20)
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```
